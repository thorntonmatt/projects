```python
import pandas as pd
bikes = pd.read_csv('/content/bikes.csv')
bikes.head()
```

Explore the data

```python
bikes.describe()
```

```python
%matplotlib inline
bikes.plot(kind='scatter', x='temperature', y='rentals')
```

```python
bikes.plot(kind='scatter', x='humidity', y='rentals')
```

```python
bikes.plot(kind='scatter', x='windspeed', y='rentals')
```

**Prepare Data**

```python
response = 'rentals'
y = bikes[(response)]
y
```

```python
predictors = list(bikes.columns)
predictors.remove(response)
x = bikes[predictors]
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1234)
```

Train the Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```
